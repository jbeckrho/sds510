```python
#importing google drive
from google.colab import drive
#importing necessary library
import pandas as pd

#reading the crime data csv from google drive
n = pd.read_csv('/content/drive/MyDrive/Phx Crime Data/crimestat (1).csv', engine='python', on_bad_lines='warn')

#cleaning the data, remove rows with missing info
clean_crime = n.dropna(subset=['ZIP', 'PREMISE TYPE', 'UCR CRIME CATEGORY'])

#organizing crimes by zip code
crimes_zip = clean_crime.groupby('ZIP').size()

#sorting zip highest to lowest
crimes_zip = crimes_zip.sort_values(ascending=False)

count = 0
print("ZIP CODE")
for zip_code, crime_count in crimes_zip.items():
    print(f"Zip code {zip_code}: {crime_count} crimes")
    count += 1
    if count >= 15:
        break

#line break
print("\n")

#looking at crime location
crimes_location = clean_crime.groupby('PREMISE TYPE').size()

#sort highest to lowest
crimes_location = crimes_location.sort_values(ascending=False)

count = 0
print("CRIME LOCATION")
for location, crime_count in crimes_location.items():

    print(f"{location}: {crime_count} crimes")
    count += 1
    if count >= 15:
        break

#organizing crimes by crime type
crime_types = clean_crime['UCR CRIME CATEGORY'].value_counts()

#line break
print("\n")

print("CRIME TYPE")
for crime_type, count in crime_types.items():
    print(f"{crime_type}: {count} crimes")
```
